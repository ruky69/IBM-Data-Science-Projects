```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
file = r"C:\Users\efere\Downloads\Topic_Survey_Assignment.csv"
df = pd.read_csv(file, index_col=0)
df.head()
```

```python
df.sort_values(by='Very interested', ascending = False, inplace = True)
df
```

```python
df = (100 * df/2233).round(2)
df
```

```python
%matplotlib inline
ax = df.plot(kind='bar', figsize=(20,8), color=['#5cb85c','#5bc0de','#d9534f'], width=0.8)
ax.set_title('The percentage of respondent in a data science survey', fontsize=16)
plt.legend(fontsize=14)
plt.xticks(fontsize=14)
plt.yticks([])
for bar in ax.patches:
    width, height = bar.get_width(), bar.get_height()
    x, y = bar.get_xy()
    ax.annotate("{}%".format(height), (x,y+height+1))
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
